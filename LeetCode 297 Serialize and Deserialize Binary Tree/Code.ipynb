```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """

        # base case
        if not root:
            return ""

        Queue = collections.deque([root])
        Flatten = []

        while Queue:
            cur = Queue.pop()
            
            if cur:
                Flatten.append(str(cur.val))
                Queue.appendleft(cur.left)
                Queue.appendleft(cur.right)
            
            else:
                Flatten.append("") # None
        
        return "/".join(Flatten)

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """

        # base case
        if not data:
            return None
        
        Flatten = data.split("/")
        ans = TreeNode(Flatten[0])
        Queue = collections.deque([ans])

        ptr = 1

        while Queue:
            cur = Queue.pop()

            if ptr < len(Flatten) and Flatten[ptr]:
                cur.left = TreeNode(int(Flatten[ptr]))
                Queue.appendleft(cur.left)

            ptr += 1

            if ptr < len(Flatten) and Flatten[ptr]:
                cur.right = TreeNode(int(Flatten[ptr]))
                Queue.appendleft(cur.right)
            
            ptr += 1
        
        return ans

# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# ans = deser.deserialize(ser.serialize(root))
```
