```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Codec:

    def serialize(self, root: TreeNode) -> str:
        """Encodes a tree to a single string.
        """

        # base case
        if not root:
            return ""
        
        in_data = self.inorder(root)
        pre_data = self.preorder(root)
        return in_data + "@" + pre_data
        

    def deserialize(self, data: str) -> TreeNode:
        """Decodes your encoded data to tree.
        """

        if not data:
            return []

        in_data, pre_data = data.split("@")
        in_data = in_data.split(",")
        pre_data = pre_data.split(",")
        in_data, pre_data = list(in_data), list(pre_data)
        tree = self.buildtree(in_data[0 : len(in_data) - 1], pre_data[0 : len(pre_data) - 1])

        return tree

    def buildtree(self, inorder, preorder):

        if not inorder or not preorder:
            return None

        cur_val = preorder[0]
        idx = inorder.index(cur_val)

        root = TreeNode(int(cur_val))
        root.left = self.buildtree(inorder[0 : idx], preorder[1 : idx + 1])
        root.right = self.buildtree(inorder[idx + 1 : ], preorder[idx + 1 : ])

        return root

    def inorder(self, root):

        stack = []
        cur = root
        ans = ""

        while cur or stack:
            while cur:
                stack.append(cur)
                cur = cur.left
            
            node = stack.pop()
            ans += str(node.val) + ","

            cur = node.right
        
        return ans
    
    def preorder(self, root):

        stack = [root]
        ans = ""

        while stack:
            node = stack.pop()
            ans += str(node.val) + ","

            if node.right:
                stack.append(node.right)
            
            if node.left:
                stack.append(node.left)
        
        return ans
        

# Your Codec object will be instantiated and called as such:
# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# tree = ser.serialize(root)
# ans = deser.deserialize(tree)
# return ans
```
