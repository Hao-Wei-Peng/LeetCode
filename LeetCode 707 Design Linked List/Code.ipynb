```python
# declare a listnode structure
class ListNode:
    def __init__(self, val = 0, next = None):
        self.val = val
        self.next = next

class MyLinkedList:

    def __init__(self):
        # dummy head node
        self.head = ListNode(0)
        # count nodes of current linked list
        self.num_node = 0

    def get(self, index: int) -> int:
        # we need to check whether index in reasonable bounded
        if index < 0 or index >= self.num_node:
            return -1
        
        cur = self.head

        # we can suppose that index value is 0
        # edge case
        for _ in range(0, index + 1):
            cur = cur.next
        
        return cur.val

    def addAtHead(self, val: int) -> None:
        
        self.addAtIndex(0, val)

    def addAtTail(self, val: int) -> None:

        self.addAtIndex(self.num_node, val)        

    def addAtIndex(self, index: int, val: int) -> None:
        
        # two conditions:
        # 1. if index equal length of linked list, append node to tail
        # 2. if index greater than length of linked list, not be inserted

        if index > self.num_node:
            return 
        
        if index < 0:
            index = 0
        
        # update current total nodes of linked list
        self.num_node += 1

        cur = self.head

        # find where we need to insert new node
        for _ in range(0, index):
            cur = cur.next
        
        new_node = ListNode(val)
        temp = cur.next
        cur.next = new_node
        new_node.next = temp

    def deleteAtIndex(self, index: int) -> None:

        if index < 0 or index >= self.num_node:
            return

        # update current total nodes of linked list
        self.num_node -= 1

        cur = self.head

        # find where we need to delete node
        for _ in range(0, index):
            cur = cur.next
        
        cur.next = cur.next.next

# Your MyLinkedList object will be instantiated and called as such:
# obj = MyLinkedList()
# param_1 = obj.get(index)
# obj.addAtHead(val)
# obj.addAtTail(val)
# obj.addAtIndex(index,val)
# obj.deleteAtIndex(index)
```
